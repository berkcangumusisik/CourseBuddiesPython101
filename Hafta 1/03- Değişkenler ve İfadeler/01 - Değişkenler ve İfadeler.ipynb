# YORUM SATIRLARI
- Yorumlar önemlidir çünkü size kodunuzu bozmadan kodunuzu açıklama şansı veririz. (bu, kodun hatalarını ayıklamak için önemlidir)
- Bir yorum satırı eklemek için '#' kullanılabilir.

```python
sayi1 = int(input("1. sayıyı giriniz:"))
sayi2 = int(input("2. sayıyı giriniz:"))
print(sayi1 + sayi2)
# burada inputtan önce int kullanma nedenimiz input ifadesi her zaman string olarak alınır.
```
